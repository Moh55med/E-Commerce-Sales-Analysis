```python
import pandas as pd
import sqlalchemy
```

```python
#Load data
df = pd.read_csv('data/UsedCarsSA_Clean_EN.csv')
#Clean data
df_clean = df[df['Price']>0]
#Create database
engine = sqlalchemy.create_engine('sqlite:///data/cars.db')
#Save to SQL
df_clean.to_sql('cars',engine,if_exists='replace',index=False)
```

```python
query = """
SELECT *
FROM cars
LIMIT 5
"""

result = pd.read_sql(query, engine)
print(result)
```

```python
query = """
SELECT Make, COUNT(*) AS total_cars
FROM cars
GROUP BY Make
ORDER BY total_cars DESC
"""

result = pd.read_sql(query, engine)
print(result)
```

```python
query = """

SELECT Make, AVG(Price) AS avg_price

FROM cars

GROUP BY Make

ORDER BY avg_price DESC

"""

result = pd.read_sql(query, engine)

print(result)
```

```python
query = """

SELECT Region, MAX(Price) AS Most_expensive_car

FROM cars

GROUP BY region

ORDER BY Most_expensive_car DESC

"""

result = pd.read_sql(query, engine)

print(result)
```

```python
query = """

SELECT Fuel_Type, AVG(Mileage) AS Average_fuel

FROM cars

GROUP BY Fuel_Type

ORDER BY Average_fuel DESC

"""

result = pd.read_sql(query, engine)

print(result)
```

```python
query = """

SELECT Gear_Type, COUNT(*) AS total_cars

FROM cars

GROUP BY Gear_Type

ORDER BY total_cars DESC

"""

result = pd.read_sql(query, engine)

print(result)
```
